```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r"D:\StatTrack\raw data\vctplayer_stats.csv")
df.head()
```

```python
df.info()
```

```python
df.drop(["Unnamed: 0"], axis="columns", inplace = True)
df.head()
```

```python
df['team'].describe()
```

```python
teams = df['team'].drop_duplicates().reset_index(drop=True)
df_team = pd.DataFrame({'team_id': range(101,101+len(teams)), 'team_name': teams})
df_team
```

```python
df = (df.merge(df_team, left_on="team", right_on="team_name", how="left")
      .drop(columns=['team_name'])
      .rename(columns = {'team_id':'team1_id'}))
```

```python
df.head(10)
```

```python
df = (df.merge(df_team, left_on="opponent", right_on="team_name", how="left")
      .drop(columns=['team_name'])
      .rename(columns = {'team_id':'team2_id'}))
df.head(10)
```

```python
df_players = df[['player','team1_id','team']].drop_duplicates().reset_index(drop = True)
player_id = range(1000, 1000+len(df_players))
df_players.insert(0, column='player_id', value=player_id)
df_players.head(20)
```

```python
df_players.rename(columns={'team1_id':'team_id'})
```

```python
df_players['player'].describe()
```

```python
dupes = df_players[df_players.groupby("player")["player"].transform("count") > 1]
dupes
```
