```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r"D:\Git Repos\Stat-Track\raw data\vctplayer_stats.csv")
df.head(20)
```

```python
df.info()
```

```python
numeric_cols = ["acs", "kast%", "adr", "hs%", "fk", "fd"]

df["kast%"] = df["kast%"].str.replace("%", "", regex=False)
df["hs%"]   = df["hs%"].str.replace("%", "", regex=False)
df['rating'] = pd.to_numeric(df['rating'], errors="coerce")
for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors="coerce").astype('Int64')
```

```python
df.dropna(inplace=True)
```

```python
print(df['team'].unique())
```

```python
df = df.drop(df[df['team'].isin(['BNY','SPB'])].index)
print(df['team'].unique())
```

```python
unique_teams = df['team'].unique()
team_df = pd.DataFrame({
    'team_id': [i for i in range(101, len(unique_teams)+101)], 
    'team_name': unique_teams
})
team_df.tail(10)
```

```python
unique_players = df[['player', 'team']].drop_duplicates().reset_index(drop=True)
player_df = pd.DataFrame({
    'player_id': [i for i in range(1001, len(unique_players)+1001)], 
    'player_name': unique_players['player'],
    'team_name': unique_players['team']
})
player_df.head()
```

```python
player_df = player_df.merge(team_df, left_on='team_name', right_on='team_name', how='left')
player_df = player_df[['player_id', 'player_name', 'team_id']]
```

```python
games_df = df.drop_duplicates(subset='game_id')
games_df = games_df.reset_index(drop=True)
games_df.shape
```

```python
games_df.head()
```

```python
columns_to_keep = ['team','score_team','opponent','score_opp','win_lose','map']
games_df = games_df[columns_to_keep]
games_df.head(10)
```

```python
games_df['game_id'] = [i for i in range (10001, games_df.shape[0]+10001)]
games_df.tail(10)
```

```python
games_df['w_score'] = games_df[['score_team', 'score_opp']].max(axis=1)
games_df['l_score'] = games_df[['score_team', 'score_opp']].min(axis=1)
games_df.head(10)
```

```python
games_df['win_team'] = np.where(games_df['win_lose'] == 'team win', games_df['team'], games_df['opponent'])
games_df.drop(['win_lose','score_team','score_opp'], axis=1, inplace=True)
games_df.head(10)
```

```python
games_df = games_df.merge(team_df.rename(columns={'team_name': 'team', 'team_id': 'team1_id'}), on='team', how='left')
games_df = games_df.merge(team_df.rename(columns={'team_name': 'opponent', 'team_id': 'team2_id'}), on='opponent', how='left')
games_df = games_df.merge(team_df.rename(columns={'team_name': 'win_team', 'team_id': 'winner_id'}), on='win_team', how='left')
games_df.head()
```

```python
games_df.drop(['team','opponent','win_team'],axis=1, inplace=True)
games_df.head()
```

```python
games_df = games_df[['game_id', 'team1_id', 'team2_id', 'map', 'winner_id', 'w_score', 'l_score']]
games_df.head()
```

```python
game_ids = [x for x in games_df['game_id'] for _ in range(10)]
```
